원래 문서(모두딥 시즌2)의 `lab-08-2_xor_nn_wide_deep.ipynb`에 해당하는 내용입니다.

## Imports

```python
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'

#for reproducibility
torch.manual_seed(777)
if device == 'cuda':
  torch.cuda.manual_seed(777)
```

## Data

```python
X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]]).to(device)
Y = torch.FloatTensor([[0], [1], [1], [0]]).to(device)
```

## Model

```python
#layers
#이 부분이 이전(08-2)과 다름
linear1 = torch.nn.Linear(2, 10, bias= True)
linear2 = torch.nn.Linear(10, 10, bias= True)
linear3 = torch.nn.Linear(10, 10, bias= True)
linear4 = torch.nn.Linear(10, 1, bias= True)
sigmoid = torch.nn.Sigmoid()

#model
model = torch.nn.Sequential(linear1, sigmoid, linear2, sigmoid, linear3, sigmoid, linear4, sigmoid).to(device)

#cost function & optimizer
criterion = torch.nn.BCELoss().to(device)
optimizer = torch.optim.SGD(model.parameters(), lr=1)
```

## Training

```python
for step in range(10000):
  hypothesis = model(X)
  cost = criterion(hypothesis, Y)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if step % 100 == 0:
    print(step, cost.item())
```

## Accuracy

```python
with torch.no_grad():
  hypothesis = model(X)
  prediction = (hypothesis > 0.5).float()
  accuracy = (prediction == Y).float().mean()
  
  print("---Hypothesis---")
  print(hypothesis.detach().cpu().numpy())
  print('---Correct---')
  print(Y.detach().cpu().numpy()) #원래 문서에서는 Y 대신에 prediction을 사용했지만, Y가 더 적절하다고 생각함 
  print("---Accuracy---")
  print(accuracy.item())
```
