# Lab 8-1: XOR


## Imports

```python
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'

#for reproducibility
torch.manual_seed(777)
if device == 'cuda':
  torch.cuda.manual_seed_all(777)
```

## Data

```python
X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]]).to(device)
Y = torch.FloatTensor([[0], [1], [1], [0]]).to(device)
```

## Model
`torch.nn.Sequential`에 대한 [참고 자료](https://dororongju.tistory.com/147)

```python
linear = torch.nn.Linear(2, 1, bias= True)
sigmoid= torch.nn.Sigmoid()

model = torch.nn.Sequential(linear, sigmoid).to(device) 
```

## Cost & Optimizer

```python
criterion = torch.nn.BCELoss().to(device)
optimizer = torch.optim.SGD(model.parameters(), lr=1)
```

## Train

```python
for step in range(1000):
  hypothesis = model(X)

  cost = criterion(hypothesis, Y)
  
  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if step % 100 == 0:
    print(step, cost.item())
```

## Accuracy

```python
with torch.no_grad():
  hypothesis = model(X)

  prediction = (hypothesis >= 0.5).float() #True if hypothesis >= 0.5, else False
  accuracy = (prediction == Y).float().mean()

  print('--Hypothesis--')
  print(hypothesis.detach().cpu().numpy())
  print('---Correct---')
  print(prediction.detach().cpu().numpy())
  print('---Accuracy---')
  print(accuracy.item())
```
